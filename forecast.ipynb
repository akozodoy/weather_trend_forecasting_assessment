We have outliers in wind_kph, pressure_mb since their maximum values are impossible. Since we do not want to loose data, we will clip all outliers to the maximun and minimum possible real world values.

```python
MAX_POSSIBLE_WIND = 400
MIN_POSSIBLE_PRESSURE = 900
MIN_POSSIBLE_PRESSURE = 1000


df["wind_kph"] = df["wind_kph"].clip(upper=MAX_POSSIBLE_WIND)

df["pressure_mb"] = df["pressure_mb"].clip(lower=MIN_POSSIBLE_PRESSURE, upper=MIN_POSSIBLE_PRESSURE)
```

```python
df["last_updated"] = pd.to_datetime(df["last_updated"]) #convert to datetime


df = df.sort_values(by=["location_name", "last_updated"]) #sorting data by location and updates for forecasting

df["year"] = df["last_updated"].dt.year
df["month"] = df["last_updated"].dt.month
df["day"] = df["last_updated"].dt.day
df["day_of_week"] = df["last_updated"].dt.dayofweek

df["temp_yesterday"] = df.groupby("location_name")["temperature_celsius"].shift(1)
df["temp_3_days_ago"] = df.groupby("location_name")["temperature_celsius"].shift(3)
df["temp_week_ago"] = df.groupby("location_name")["temperature_celsius"].shift(7)            #previous days temp



df["temp_mean_3_days"] = df.groupby("location_name")["temperature_celsius"].rolling(3).mean().reset_index(0, drop=True)
df["temp_mean_week"] = (df.groupby("location_name")["temperature_celsius"].shift(1).rolling(7).mean().reset_index(0, drop=True))            #previous days aberage temp


df["target_temp"] = df.groupby("location_name")["temperature_celsius"].shift(-7)
```

## TRAIN/TEST SPLIT

```python
train_part = []
test_part = []

for laction, group in df.groupby("location_name"):
    group = group.sort_values("last_updated")
    index_to_split = int(len(group) * 0.8)

    if index_to_split == 0 or index_to_split == len(group):
        continue

    train_part.append(group.iloc[:index_to_split])
    test_part.append(group.iloc[index_to_split:])

train_df = pd.concat(train_part).sort_values(["location_name", "last_updated"])
test_df = pd.concat(test_part).sort_values(["location_name", "last_updated"])

print(train_df.shape)
print(test_df.shape)
```

```python
X_train = train_df.drop(columns=["target_temp",
                                 "location_name",
                                 "last_updated", 
                                 "feels_like_celsius", 
                                 "temperature_celsius", 
                                 "country", 
                                 "temperature_fahrenheit",
                                 "wind_mph", 
                                 "feels_like_fahrenheit",
                                 "pressure_in",
                                 "precip_in",
                                 "visibility_miles",
                                 "gust_mph",
                                 "timezone", 
                                 "last_updated_epoch",
                                 "air_quality_gb-defra-index", 
                                 "gust_mph",
                                 "moon_phase"])
y_train = train_df["target_temp"]

X_test = test_df.drop(columns=["target_temp",
                                 "location_name",
                                 "last_updated", 
                                 "feels_like_celsius", 
                                 "temperature_celsius", 
                                 "country", 
                                 "temperature_fahrenheit",
                                 "wind_mph", 
                                 "feels_like_fahrenheit",
                                 "pressure_in",
                                 "precip_in",
                                 "visibility_miles",
                                 "gust_mph",
                                 "timezone", 
                                 "last_updated_epoch",
                                 "air_quality_gb-defra-index", 
                                 "gust_mph",
                                 "moon_phase"])
y_test = test_df["target_temp"]
```

```python
X_train.columns.tolist()
```

```python
numeric_features = [
    "latitude",
    "longitude",
    "wind_kph",
    "wind_degree",
    "pressure_mb",
    "precip_mm",
    "humidity",
    "cloud",
    "visibility_km",
    "uv_index",
    "gust_kph",
    "air_quality_Carbon_Monoxide",
    "air_quality_Ozone",
    "air_quality_Nitrogen_dioxide",
    "air_quality_Sulphur_dioxide",
    "air_quality_PM2.5",
    "air_quality_PM10",
    "air_quality_us-epa-index",
    "moon_illumination",
    "year",
    "month",
    "day",
    "temp_yesterday",
    "temp_3_days_ago",
    "temp_week_ago",
    "temp_mean_3_days",
    "temp_mean_week",
]

categorical_features = [
    "condition_text",
    "wind_direction",
    "sunrise",
    "sunset",
    "moonrise",
    "moonset"
]
```

## LINEAR MODEL

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression

preprocessor = ColumnTransformer(
    transformers= [("num", StandardScaler(), numeric_features),
                   ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)]
)

lr_pipeline = Pipeline(steps=[("preprocessor", preprocessor), 
                              ("model", LinearRegression())])
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

tscv = TimeSeriesSplit(n_splits=3)

cv_results = []
all_y_true = []
all_y_pred = []
for location, group in train_df.groupby("location_name"):
    group = group.sort_values("last_updated").copy()

    if len(group) < 20:
        continue

    X_loc = group[numeric_features + categorical_features]
    y_loc = group["target_temp"]

    cv_mae = []
    cv_rmse = []
    cv_r2 = []

    for train_idx, val_idx in tscv.split(X_loc):
        X_train_fold = X_loc.iloc[train_idx]
        X_val_fold = X_loc.iloc[val_idx]
        y_train_fold = y_loc.iloc[train_idx]
        y_val_fold = y_loc.iloc[val_idx]

        lr_pipeline.fit(X_train_fold, y_train_fold)
        pred = lr_pipeline.predict(X_val_fold)

        all_y_true.extend(y_val_fold.tolist())
        all_y_pred.extend(pred.tolist())


print("GLOBAL CV MAE:", mean_absolute_error(all_y_true, all_y_pred))
print("GLOBAL CV RMSE:", np.sqrt(mean_squared_error(all_y_true, all_y_pred)))
print("GLOBAL CV R2:", r2_score(all_y_true, all_y_pred))
```

```python
lr_pipeline.fit(X_train, y_train)
lr_pred = lr_pipeline.predict(X_test)

print("LR MAE:", mean_absolute_error(y_test, lr_pred))
print("LR RMSE:", np.sqrt(mean_squared_error(y_test, lr_pred)))
print("LR R2:", r2_score(y_test, lr_pred))
```

## RIDGE

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

ridge_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", Ridge())
])

ridge_params = {
    "model__alpha": [0.1, 1.0, 10.0, 50.0, 100.0]
}

ridge_search = GridSearchCV(
    ridge_pipeline,
    ridge_params,
    cv=tscv,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)

ridge_search.fit(X_train, y_train)

print("Best Ridge:", ridge_search.best_params_)
print("Best CV RMSE:", -ridge_search.best_score_)
```

```python
ridge_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", Ridge(
        alpha=100
    ))
])

ridge_pipeline.fit(X_train, y_train)
pred_ridge = ridge_pipeline.predict(X_test)

print("Ridge R2:", r2_score(y_test, pred_ridge))
print("Ridge RMSE:", np.sqrt(mean_squared_error(y_test, pred_ridge)))
print("Ridge MAE:", mean_absolute_error(y_test, pred_ridge))
```
